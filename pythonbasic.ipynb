3.  Compare and contrast mutable and immutable objects in Python with examples.

# -> Mutable object

*   It can be modified after creation
  
  ex: list,Dictionaries
*  if object modified Same object is updated.

 # Immutable object

*   It content cannot be changed after creation.
  
  ex: tuple,string, integer
*  if object modified creates a new object in memory.

```
# Mutable
a = [1, 2, 3]
b = a
b.append(4)
print(a)  # Output: [1, 2, 3, 4] — both a and b point to the same object

# Immutable
x = "hello"
y = x
y = y + " world"
print(x)  # Output: "hello" — x remains unchanged
print(y)  # Output: "hello world"
```

# *   Comparison (Relational) Operators
Used to compare two values such as equal to, greater than , less than, not equal to etc.


```
# Define two numbers
x = 10
y = 5

# Equal to
print("x == y:", x == y)   # False

# Not equal to
print("x != y:", x != y)   # True

print("x > y:", x > y)     # True

# Less than

# Less than or equal to
print("x <= y:", x <= y)   # False
```


# *   Assignment Operators
Used to assign values to variables.


```
# Start with a value
x = 10
print("Initial x:", x)

# Add and assign
x += 5   # x = x + 5
print("After x += 5:", x)

# Subtract and assign
x -= 3   # x = x - 3
print("After x -= 3:", x)

# Multiply and assign
x *= 2   # x = x * 2
print("After x *= 2:", x)

# Divide and assign
x /= 4   # x = x / 4

# Modulus and assign

# Exponentiate and assign
x **= 3  # x = x ** 3
print("After x **= 3:", x)
```


# *  Logical Operators
Used to combine conditional statements.


```
# Define two conditions
a = 10
b = 5

# Logical AND: True if both conditions are True
print("a > 5 and b < 10:", a > 5 and b < 10)  # True and True → True

# Logical OR: True if at least one condition is True
print("a < 5 or b < 10:", a < 5 or b < 10)    # False or True → True

# Logical NOT: Reverses the result
print("not(a > b):", not(a > b))              # not(True) → False
```



#*   Bitwise Operators
Operate on bits of integers.


```
# Define two numbers
a = 5     # In binary: 0101
b = 3     # In binary: 0011

# Bitwise AND
print("a & b:", a & b)     # 0101 & 0011 = 0001 → 1

# Bitwise OR
print("a | b:", a | b)     # 0101 | 0011 = 0111 → 7

# Bitwise XOR (exclusive OR)
print("a ^ b:", a ^ b)     # 0101 ^ 0011 = 0110 → 6

# Bitwise NOT
print("~a:", ~a)           # ~0101 = -(0101 + 1) = -6 (in two's complement)

# Left shift
print("a << 1:", a << 1)   # 0101 << 1 = 1010 → 10

# Right shift
print("a >> 1:", a >> 1)   # 0101 >> 1 = 0010 → 2

```












5. Explain the concept of type casting in Python with examples.
    
    -> Type casting is the process of converting one data type to another. It allows you to transform variables of one type (like int, str, or float) into another type so they can be used in different ways.

*   Python provides built-in functions for type casting, such as int(), float(), str(), list(), tuple(), dict(), etc.
*   Type casting is often necessary for performing arithmetic operations, data manipulation, and input/output operations in Python programs.

*   There are two main types of casting in Python:

1.  Implicit Type Casting (automatic)
2.   Explicit Type Casting (manual)

````python
int_val = 10
float_val = 3.14
str_num = "50"
str_float = "9.81"
bool_val = True
list_val = [1, 2, 3]
tuple_val = (4, 5, 6)

# Integer to Float
int_to_float = float(int_val)
print("Integer to Float:", int_to_float)

# Float to Integer
float_to_int = int(float_val)
print("Float to Integer:", float_to_int)

# String to Integer
str_to_int = int(str_num)
print("String to Integer:", str_to_int)

# String to Float
str_to_float = float(str_float)
print("String to Float:", str_to_float)

# Integer to String
int_to_str = str(int_val)
print("Integer to String:", int_to_str)

# Float to String
float_to_str = str(float_val)
print("Float to String:", float_to_str)

# Boolean to Integer
bool_to_int = int(bool_val)
print("Boolean to Integer:", bool_to_int)

# Integer to Boolean
int_to_bool = bool(0)
print("Integer to Boolean (0):", int_to_bool)

# List to Tuple
list_to_tuple = tuple(list_val)
print("List to Tuple:", list_to_tuple)

# Tuple to List
tuple_to_list = list(tuple_val)
print("Tuple to List:", tuple_to_list)

# List to String (via str)
list_to_str = str(list_val)
print("List to String:", list_to_str)
````

6. How do conditional statements work in Python? Illustrate with examples.
   -> Conditionals in Python are used to execute code based on the evaluation of one or more conditions.

````python
#if Statement
age = 18
if age >= 18:
    print("You are an adult.")

#if-else Statement
age = 16
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")

#if-elif-else Statement
marks = 85
if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
else:
    print("Grade: F")

#nested Statement
age = 25
citizen = True
if age >= 18:
    if citizen:
        print("Eligible to vote.")
    else:
        print("Not a citizen.")
else:
    print("Too young to vote.")
````
